```python
cam = cv2.VideoCapture(0)
count=0
strp="wait"
text_x=200
text_y=200
text=""
while True:
    count=count+1
    ret,frame = cam.read()
    ret2,frame2 = cam.read()
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    black=np.zeros((480,640,3))
    black[:] = 0
    frame2= black
    
    if (ret==False):
        continue
    
    cv2.rectangle(frame,(0,61),(232,293),(255,0,0),4)
    cut_frame=frame[65:289,4:228,:]
    
    if(count>=30):
        count=0
        x = np.reshape(cut_frame,[1,224,224,3])
        x = preprocess_input(x)
        pred=model_load.predict(x)
        pred_i = np.argmax(pred)
        if(pred[0,pred_i]>0.85):
            strp = pred_image(pred_i)
            if(strp=="del"):
                text=text[:-1]
            if (strp=="nothing"):
                text=text
            if (strp=="space"):
                text=text+" "
            else:
                text=text+strp
    cv2.putText(frame2, strp, (400,50), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0),2)
    cv2.putText(frame2, text, (200,200), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0),2)
    cv2.imshow("Image",frame)
    cv2.imshow("Image2",frame2)
    key_pressed = cv2.waitKey(1) & 0xFF
    if(key_pressed==ord('a')):
        break
cam.release()
cv2.destroyAllWindows()
```

```python
# text to voice
import pyttsx3

engine = pyttsx3.init() 
  
# testing 
engine.say(text)

engine.runAndWait()
```
